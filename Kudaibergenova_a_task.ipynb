```python
#A.1
import numpy as np
resources_matrix = np.array([
    [10, 20],
    [15, 25],
    [20, 30]
])
allocation_factors = np.array([
    [0.8, 0.2],
    [0.1, 0.9]
])
result = np.dot(resources_matrix, allocation_factors)
print("result matrix")
print(result)
```

```python
#A.3
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
forecast_values = np.array([-2, -1, 0, 1, 2])
result = sigmoid(forecast_values)
print("Forecast values:", forecast_values)
print("Sigmoid output:", result)
```

```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_gradient(x):
    s = sigmoid(x)
    return s * (1 - s)
test_values = np.array([-2, -1, 0, 1, 2])
grad_output = sigmoid_gradient(test_values)
print("Input values:", test_values)
print("Sigmoid gradient:", grad_output)
```
